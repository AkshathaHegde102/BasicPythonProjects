## Find PI to the Nth Digit

Enter a number and have the program generate PI up to that many decimal places. Keep a limit to how far the program will go.

```python
from decimal import *
```

```python
def find_pi(number):
    try:
        number = int(number) + 1
    except:
        return TypeError
    if number > 100:
        print("limit exceeded")
        number = 100
    getcontext().prec = number
    pi = Decimal(22)/Decimal(7)
    return pi
```

```python
x = input('Enter the decimal place number as an integer (limit of 100):')
find_pi(x)
```
